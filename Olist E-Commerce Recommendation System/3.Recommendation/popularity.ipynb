# Popularity Recommendation

```python
import os
import numpy as np
import pandas as pd
import snowflake.connector
```

Importing the data from Snowflake to a dataframe.

Masking the credentials

```python
ctx = snowflake.connector.connect(
user='******',
password='******',
account='******'
)
query = "SELECT PR.PRODUCT_ID,O.ORDER_ID,O.CUSTOMER_ID, PR.PRODUCT_CATEGORY_NAME, OP.PAYMENT_VALUE,O.ORDER_DELIVERED_CUSTOMER_DATE,OC.CUSTOMER_UNIQUE_ID,ORD.REVIEW_SCORE, OC.CUSTOMER_CITY FROM OLIST.PUBLIC.OLIST_PRODUCTS_DATASET AS PR JOIN OLIST.PUBLIC.OLIST_ORDER_ITEMS_DATASET AS OI ON PR.PRODUCT_ID = OI.PRODUCT_ID JOIN OLIST.PUBLIC.OLIST_ORDERS_DATASET O ON OI.ORDER_ID = O.ORDER_ID JOIN OLIST.PUBLIC.OLIST_CUSTOMERS_DATASET OC ON O.CUSTOMER_ID = OC.CUSTOMER_ID JOIN OLIST.PUBLIC.OLIST_ORDER_PAYMENTS_DATASET OP ON O.ORDER_ID = OP.ORDER_ID JOIN OLIST.PUBLIC.OLIST_ORDER_REVIEWS_DATASET ORD ON O.ORDER_ID = ORD.ORDER_ID WHERE O.ORDER_STATUS = 'delivered'"
df = pd.read_sql(query, ctx)
df
```

```python
df.PRODUCT_CATEGORY_NAME.unique()
```

```python
df['PRODUCT_CATEGORY_NAME'].replace({None:"None"}, inplace = True)
```

```python
df.columns
```

```python
df.head()
```

```python
df['Product_Price']=df['PAYMENT_VALUE']
```

Adding a dataframe containing Product and Price

```python
df_price=df[['PRODUCT_ID','Product_Price']].groupby(['PRODUCT_ID']).agg('mean')
```

```python
df_price
```

```python
df_category=df.groupby(["PRODUCT_ID",'PRODUCT_CATEGORY_NAME']).count().reset_index()[['PRODUCT_ID','PRODUCT_CATEGORY_NAME']]
```

```python
df_category
```

Calculating the frequency

```python
df_frequent=df[['PRODUCT_ID','CUSTOMER_ID']].groupby(['PRODUCT_ID']).agg('count')
```

```python
df_frequent.columns=['Count']
```

```python
df_frequent
```

Calculating the avg review score

```python
df_mostrated=df[['PRODUCT_ID','REVIEW_SCORE']].groupby(['PRODUCT_ID']).agg('mean')
```

```python
df_mostrated
```

Merging the dataframes

```python
df_popular=pd.merge(df_frequent,df_mostrated,left_index=True, right_index=True)
df_popular
```

```python
df_popular=pd.merge(df_popular,df_price,left_index=True, right_index=True)
df_popular
```

```python
df_popular
```

```python
df_popular=df_popular.sort_values(by=['Count','REVIEW_SCORE'], ascending=False)
```
